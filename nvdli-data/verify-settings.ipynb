```python
import torch
import sys
import time

print("=" * 50)
print("🚀 JETSON ORIN NANO SYSTEM VERIFICATION")
print("=" * 50)

# 1. Environment Configurations
print(f"🔹 Python Version: {sys.version.split()[0]}")
print(f"🔹 PyTorch Version: {torch.__version__}")

# 2. CUDA & GPU Acceleration Verification
cuda_available = torch.cuda.is_available()
print(f"🔹 CUDA GPU Acceleration Available: {cuda_available}")

if cuda_available:
    device_id = torch.cuda.current_device()
    device_name = torch.cuda.get_device_name(device_id)
    print(f"🔹 Target Device Name: {device_name}")
    print(f"🔹 Compute Capability: {torch.cuda.get_device_capability(device_id)}")
    
    # 3. Execution Stress Test (Matrix Multiplication Tensor Test)
    print("\n⏳ Running GPU Tensor Stress Test...")
    start_time = time.time()
    
    # Create two massive random tensors directly on the Jetson's GPU memory
    X = torch.rand(4000, 4000, device="cuda")
    Y = torch.rand(4000, 4000, device="cuda")
    
    # Perform matrix multiplication (Fully accelerated via Tensor Cores)
    Z = torch.matmul(X, Y)
    
    # Sync CUDA threads to get accurate timing calculations
    torch.cuda.synchronize()
    elapsed_time = time.time() - start_time
    
    print("✅ Tensor Core Stress Test Complete!")
    print(f"⏱️  Execution Speed for (4000x4000) Matrix Mult: {elapsed_time:.4f} seconds")
    print(f"📦 Result Matrix Tensor Shape: {Z.shape}")
else:
    print("\n❌ ERROR: CUDA is not initialized correctly. Check your container runtime configurations.")

print("=" * 50)
```
